```python
import gradio as gr
import tensorflow as tf
import numpy as np
```

```python
from fastai.vision import *
import torchvision.transforms as T
```

```python
imp = gr.inputs.Image(shape=(224, 224))
```

```python
#Function to use in gradio interface
def grad(imp):
  lis=[]
  img_tensor = T.ToTensor()(imp)
  img = Image(img_tensor)
  prediction1 = learn1.predict(img)
  prediction2 = learn2.predict(img)
  prediction3 = learn3.predict(img)
  #print(prediction)
  #print(prediction1[0])
  #print(prediction2[0])
  #print(prediction3[0])
  lis.append(prediction1[2].tolist())
  lis.append(prediction2[2].tolist())
  lis.append(prediction3[2].tolist())
  #print(lis)
  lis=flatten(lis)
  #print(lis)
  cls=max(lis)
  if cls==lis[2]:
    lis[2]=lis[2]-0.35
    ans=max(lis)
  else:
    ans=max(lis)
    #print(ans)
  na1="The probability of COVID19  ="+str(lis[0])
  na2="The probability of non covid19 condition ="+str(lis[1])
  na3="The probability of Adenocarcinoma  ="+str(lis[2])
  na4="The probability of large call caarcinoma  ="+str(lis[3])
  na5="The probability of non cancer condition ="+str(lis[4])
  na6="The probability of Squamous cell carcinoma  ="+str(lis[5])
  na7="The probability of non tuberculosis condition  ="+str(lis[6])
  na8="The probability of Tuberculosis  ="+str(lis[7])
  if ans==lis[0]:
    return na1,na2,na3,na4,na5,na6,na7,na8,"FINAL OUTPUT:It is COVID 19 image"
  elif ans==lis[1]:
    return na1,na2,na3,na4,na5,na6,na7,na8,"FINAL OUTPUT:It is NOT a COVID 19 image, it is NORMAL image"
  elif ans==lis[2]:
    return na1,na2,na3,na4,na5,na6,na7,na8,"FINAL OUTPUT:It is ADENOCARCINOMA image"
  elif ans==lis[3]:
    return na1,na2,na3,na4,na5,na6,na7,na8,"FINAL OUTPUT:It is LARGE CELL CARCINOMA image"
  elif ans==lis[4]:
    return na1,na2,na3,na4,na5,na6,na7,na8,"FINAL OUTPUT:It is NOT A CANCER image, it is NORMAL image"
  elif ans==lis[5]:
    return na1,na2,na3,na4,na5,na6,na7,na8,"FINAL OUTPUT:It is SQUAMOUS CELL CARCINOMA image"
  elif ans==lis[6]:
    return na1,na2,na3,na4,na5,na6,na7,na8,"FINAL OUTPUT:It is  NOT A TUBERCULOSIS image, it is NORMAL image"
  else:
    return na1,na2,na3,na4,na5,na6,na7,na8,"FINAL OUTPUT:It is TUBERCULOSIS"

```

```python
gr.Interface(fn=grad, inputs=imp, outputs=['text','text','text','text','text','text','text','text','text'], capture_session=True).launch(debug=True)
```
